```python
import logging
import requests
from bs4 import BeautifulSoup

def get_text(url):
    ''' 
    scrape the text from the url specified
    '''
    text=''
    page = requests.get(url).text
    soup = BeautifulSoup(page, "lxml")
    try:
        for element in soup.find_all(class_='css-1i0edl6 e2kc3sl0'):
            text=text+' '+element.text
    except:
        msg = 'Failed to get scrape text from page '+url
        logging.error(msg)
        return('')
    return(text)
```

```python
# parser for newer aticles that seem to be different

def get_text_type2(url):
    text=''
    page = requests.get(url).text
    soup = BeautifulSoup(page, "lxml")

    for element in soup.find_all(class_='story-body-text story-content'):
        text=text+' '+element.text

    return(text)
```

```python
# parser for older articles

def get_text_type3(url):
    text=''
    page = requests.get(url).text
    soup = BeautifulSoup(page, "lxml")

    for element in soup.find_all(class_='articleBody'):
        text=text+' '+element.text

    return(text)
```
